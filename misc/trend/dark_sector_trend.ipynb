```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import os
import sys
import copy 

from astropy.cosmology import FlatLambdaCDM
from astropy.table import Table, Column, vstack, join

import numpy as np 

import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({
    "text.usetex": False,
    "font.family": "Heiti TC",
    "font.sans-serif": ["Heiti TC"]})
```

```python
dark_matter = Table.read('dark_matter.csv', format='csv')
dark_energy = Table.read('dark_energy.csv', format='csv')

h0_tension = Table.read('hubble_tension.csv', format='csv')
cosmos_tension = Table.read('cosmology_tension.csv', format='csv')
lensing_tension = Table.read('lensing_tension.csv', format='csv')
```

```python
plt.figure(figsize=(8, 8))
plt.xlabel('', fontsize=30);
plt.ylabel('', fontsize=30);

plt.plot(dark_matter['Year'], dark_matter['Ref Count'] / np.max(dark_matter['Ref Count']), label=u'',
         linewidth=7, alpha=0.5, linestyle='--', c='k')
plt.plot(dark_energy['Year'], dark_energy['Ref Count'] / np.max(dark_energy['Ref Count']), label=u'',
         linewidth=7, alpha=0.5, linestyle='-.', c='olivedrab')

plt.plot(cosmos_tension['Year'], cosmos_tension['Ref Count'] / np.max(cosmos_tension['Ref Count']), 
         label=u'', linewidth=6, alpha=0.8, linestyle='-', c='darkorange')
plt.plot(h0_tension['Year'], h0_tension['Ref Count'] / np.max(h0_tension['Ref Count']), 
         label=u'', linewidth=6, alpha=0.8, linestyle='-', c='dodgerblue')
plt.plot(lensing_tension['Year'], lensing_tension['Ref Count'] / np.max(lensing_tension['Ref Count']), 
         label=u'', linewidth=6, alpha=0.9, linestyle='-', c='darkviolet')

plt.legend(loc='best', fontsize=24)

plt.axvline(2011, linestyle='--', linewidth=4, color='grey', zorder=0, alpha=0.8)
plt.axvline(2019, linestyle='--', linewidth=4, color='grey', zorder=0, alpha=0.8)

plt.tight_layout()
plt.savefig('/Users/song/Desktop/dark_sector_1.png', dpi=200)
```

```python
cdm = Table.read('cold_dark_matter.csv', format='csv')
wdm = Table.read('warm_dark_matter.csv', format='csv')

sidm = Table.read('self_interacting_dark_matter.csv', format='csv')
fdm = Table.read('fuzzy_dark_matter.csv', format='csv')
pbh = Table.read('primordial_black_hole.csv', format='csv')
axion = Table.read('axion.csv', format='csv')
```

```python
plt.figure(figsize=(8, 8))
plt.xlabel('', fontsize=30);
plt.ylabel('', fontsize=30);

plt.plot(cdm['Year'], cdm['Ref Count'] / np.max(cdm['Ref Count']), label=u'',
         linewidth=7, alpha=0.5, linestyle='--', c='k')
plt.plot(wdm['Year'], wdm['Ref Count'] / np.max(wdm['Ref Count']), label=u'',
         linewidth=7, alpha=0.5, linestyle='-.', c='olivedrab')

plt.plot(sidm['Year'], sidm['Ref Count'] / np.max(sidm['Ref Count']), 
         label=u'', linewidth=6, alpha=0.8, linestyle='-', c='darkorange')
plt.plot(fdm['Year'], fdm['Ref Count'] / np.max(fdm['Ref Count']), 
         label=u'', linewidth=6, alpha=0.8, linestyle='-', c='dodgerblue')
plt.plot(axion['Year'], axion['Ref Count'] / np.max(axion['Ref Count']), 
         label=u'', linewidth=6, alpha=0.9, linestyle='-', c='darkviolet')
plt.plot(pbh['Year'], pbh['Ref Count'] / np.max(pbh['Ref Count']), 
         label=u'', linewidth=6, alpha=0.9, linestyle='-', c='green')

plt.legend(loc='best', fontsize=23)

plt.axvline(2011, linestyle='--', linewidth=4, color='grey', zorder=0, alpha=0.8)
plt.axvline(2019, linestyle='--', linewidth=4, color='grey', zorder=0, alpha=0.8)

plt.tight_layout()
plt.savefig('/Users/song/Desktop/dark_sector_2.png', dpi=200)
```
