```python
tab_sh_pos=np.vstack([tab_sh['x']/(1+z_tab[72])/(0.6774),tab_sh['y']/(1+z_tab[72])/(0.6774),tab_sh['z']/(1+z_tab[72])/(0.6774)]).T
```

```python
def dist3D(central,sate,period_box=205*1e3/1.4/0.6774):
    d=central-sate
    d=d - period_box*np.round(d/period_box)
    dist=np.sqrt(np.sum(d**2,axis=-1))
    return dist
```

```python
with open(file_dir+'galaxies_tng100_072_central_HaloPos.txt','rb') as f:
    halo_pos_tab=pickle.load(f)
```

```python
i=76#let it from 0 to 235 is all
L=1000 #sphere radius in kpc to select around satellite
#halo=il.groupcat.loadSingle(basePath, 72, haloID=tng100_cen_HaloID[i])
halo_pos=halo_pos_tab[i]
#R200=halo['Group_R_Crit200']/(1+z_tab[72])/(0.6774)
mask_pos=(dist3D(tab_sh_pos,halo_pos)<L)&(dist3D(tab_sh_pos,halo_pos)>1)
mask_mass=(tab_sh['StellarMass']/0.6774>10**9)
tab_sh_0=Table(tab_sh[mask_pos&mask_mass])
tab_sh_0['x']=tab_sh_0['x']/(1+z_tab[72])/(0.6774)-halo_pos[0]
tab_sh_0['y']=tab_sh_0['y']/(1+z_tab[72])/(0.6774)-halo_pos[1]
tab_sh_0['z']=tab_sh_0['z']/(1+z_tab[72])/(0.6774)-halo_pos[2]
```
